## **Buat Proses untuk Memproses Dataset**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def plot_population_data(file_path):
    # Membaca dataset dari file yang telah diunggah
    table = pd.read_csv(file_path)
    table.head()

    x_label = table['NAMA KELURAHAN']
    plt.bar(x=np.arange(len(x_label)), height=table['LAKI-LAKI WNI'])
    plt.xticks(np.arange(len(x_label)), table['NAMA KELURAHAN'], rotation=90)
    plt.xlabel('Kelurahan di Jakarta Utara')
    plt.ylabel('Jumlah Penduduk Laki - Laki')
    plt.title('Persebaran Jumlah Penduduk Laki- Laki di Jakarta Utara')

    plt.show()
```

## **Panggil Proses yang Telah Dibuat**

Jika ingin meng-custom dataset ini, pastikan untuk mengganti `'nama_file.csv'` dengan nama file dataset yang ingin digunakan atau yang di unggah ke Google Colab. Dan memplot data dengan fungsi `plot_population_data`.

```python
# Mengganti 'nama_file.csv' dengan nama file yang tepat setelah diunggah
nama_file = 'data-jumlah-penduduk-berdasarkan-jenis-kelamin-tahun-2021-jakarta-utara.csv'

# Memanggil prosedur untuk memplot data
plot_population_data(nama_file)
```
